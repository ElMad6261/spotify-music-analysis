```python
import pandas as pd
import numpy as np

DATA_PATH = '/content/drive/MyDrive/spotify-analysis/data/raw/dataset.csv'

df = pd.read_csv(DATA_PATH)

print("Shape:", df.shape)
print("\nColumnas:", df.columns.tolist())
print("\nValores nulos:\n", df.isnull().sum())
print("\nDuplicados:", df.duplicated().sum())

df.head()
```

```python
# 1. Eliminar columna inútil
df = df.drop(columns=['Unnamed: 0'])

# 2. Eliminar las 3 filas con nulos (son solo 1 por columna)
df = df.dropna(subset=['artists', 'album_name', 'track_name'])

# 3. Convertir duration_ms a minutos (más legible)
df['duration_min'] = (df['duration_ms'] / 60000).round(2)
df = df.drop(columns=['duration_ms'])

# 4. Convertir explicit a int (True/False → 1/0)
df['explicit'] = df['explicit'].astype(int)

print("Shape final:", df.shape)
print("Nulos restantes:", df.isnull().sum().sum())
df.head()
```

**LIMPIAR DATOS**

```python
# Recargar el dataset limpio desde el principio
df = pd.read_csv('/content/drive/MyDrive/spotify-analysis/data/raw/dataset.csv')

# Eliminar columnas innecesarias si existen
cols_to_drop = ['Unnamed: 0']
df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

# Eliminar filas con nulos
df = df.dropna(subset=['artists', 'album_name', 'track_name'])

# Convertir duration_ms a minutos (solo si existe)
if 'duration_ms' in df.columns:
    df['duration_min'] = (df['duration_ms'] / 60000).round(2)
    df = df.drop(columns=['duration_ms'])

# Convertir explicit a int
df['explicit'] = df['explicit'].astype(int)

print("Shape final:", df.shape)
print("Columnas:", df.columns.tolist())
print("Nulos restantes:", df.isnull().sum().sum())
df.head()
```

```python
# Features derivadas que añaden valor al análisis
df['energy_dance_score'] = (df['energy'] * df['danceability']).round(4)
df['acoustic_vs_energy']  = (df['acousticness'] - df['energy']).round(4)

print("Columnas finales:", df.columns.tolist())
```

```python
PROCESSED_PATH = '/content/drive/MyDrive/spotify-analysis/data/processed/dataset_clean.csv'

import os
os.makedirs('/content/drive/MyDrive/spotify-analysis/data/processed', exist_ok=True)

df.to_csv(PROCESSED_PATH, index=False)
print("Guardado en:", PROCESSED_PATH)
```
